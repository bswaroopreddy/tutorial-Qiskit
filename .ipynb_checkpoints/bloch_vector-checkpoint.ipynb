```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
from qiskit.quantum_info import Operator, Statevector, DensityMatrix
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram, plot_bloch_vector
import numpy as np
import matplotlib.pyplot as plt
from qiskit.visualization.bloch import Bloch


# Pauli matrices
X = np.array([[0, 1], [1, 0]])
Y = np.array([[0, -1j], [1j, 0]])
Z = np.array([[1, 0], [0, -1]])

# Define a pure state |+> = (|0 + |1)/2
#psi = Statevector([1/np.sqrt(2), 1/np.sqrt(2)])
# psi = Statevector([1, 0])

global_phase = np.exp(6j)
vec = np.array([4j/5, 3/5])
psi = Statevector(global_phase * vec)

# Convert to densoty matrix
rho = DensityMatrix(psi).data

# Compute Bloch components manually
x = np.trace(rho @ X).real
y = np.trace(rho @ Y).real
z = np.trace(rho @ Z).real
bloch_vector = [x, y, z]
print(bloch_vector)

# Create Bloch sphere manually
b = Bloch()
b.add_vectors(bloch_vector)
b.title = "Bloch Sphere: |+"
b.show()
```

```python
# Block-vector for a mixed state rho = 0.6 |0><0| + 0.4 |1><1|
rho = np.matrix([[0.6, 0], [0,0.4]])

# Compute Bloch components manually
x = np.trace(rho @ X).real
y = np.trace(rho @ Y).real
z = np.trace(rho @ Z).real
bloch_vector = [x, y, z]
print(bloch_vector)

# Create Bloch sphere manually
b = Bloch()
b.add_vectors(bloch_vector)
b.title = "Bloch Sphere: |+"
b.show()
```
