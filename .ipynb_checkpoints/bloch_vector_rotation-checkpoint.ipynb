```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
from qiskit.quantum_info import Operator, Statevector, DensityMatrix
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram, plot_bloch_vector
import numpy as np
import matplotlib.pyplot as plt
from qiskit.visualization.bloch import Bloch

# Initial state: |+⟩ = (|0⟩ + |1⟩)/√2
psi = np.array([[1/np.sqrt(2)], [1/np.sqrt(2)]])
#psi = np.array([[np.sqrt(3)/2], [1/2]])

rho = psi @ psi.conj().T
print(psi)
print(rho)

# Pauli matrices
X = np.array([[0, 1], [1, 0]])
Y = np.array([[0, -1j], [1j, 0]])
Z = np.array([[1, 0], [0, -1]])

# Compute Bloch components manually
x = np.trace(rho @ X).real
y = np.trace(rho @ Y).real
z = np.trace(rho @ Z).real
bloch_vector = [x, y, z]
print(bloch_vector)

# Create Bloch sphere manually
b = Bloch()
b.add_vectors(bloch_vector)
b.title = "Bloch vector before rotation "
b.show()
```

```python
# Apply unitary: Rz(π/2)
phi = np.pi / 2
U = np.array([[np.exp(-1j*phi/2), 0],
              [0, np.exp(1j*phi/2)]])
print(U)

rho_rot = U @ rho @ U.conj().T
# Bloch vector after rotation
x = np.trace(rho_rot @ X).real
y = np.trace(rho_rot @ Y).real
z = np.trace(rho_rot @ Z).real

bloch_vector = [x, y, z]
print(bloch_vector)

# Create Bloch sphere manually
b = Bloch()
b.add_vectors(bloch_vector)
b.title = "Bloch vector after rotation about z-axis"
b.show()
```

```python
# Apply unitary: Rx(π/2)
phi = np.pi / 2
U = np.array([[np.cos(phi/2), -1j * np.sin(phi/2)],
              [-1j * np.sin(phi/2), np.cos(phi/2)]])
print(U)

rho_rot = U @ rho @ U.conj().T
# Bloch vector after rotation
x = np.trace(rho_rot @ X).real
y = np.trace(rho_rot @ Y).real
z = np.trace(rho_rot @ Z).real

bloch_vector = [x, y, z]
print(bloch_vector)

# Create Bloch sphere manually
b = Bloch()
b.add_vectors(bloch_vector)
b.title = "Bloch vector after rotation about x-axis"
b.show()
```

```python
# Apply unitary: Ry(π/2)
phi = np.pi / 2
U = np.array([[np.cos(phi/2), -np.sin(phi/2)],
              [np.sin(phi/2), np.cos(phi/2)]])
print(U)

rho_rot = U @ rho @ U.conj().T
# Bloch vector after rotation
x = np.trace(rho_rot @ X).real
y = np.trace(rho_rot @ Y).real
z = np.trace(rho_rot @ Z).real

bloch_vector = [x, y, z]
print(bloch_vector)

# Create Bloch sphere manually
b = Bloch()
b.add_vectors(bloch_vector)
b.title = "Bloch vector after rotation about y-axis"
b.show()
```

```python
# Apply unitary: Ry(π/2) and Rz(π/2)
#psi = np.array([1 / np.sqrt(2), 1 / np.sqrt(2)])
psi = np.array([np.sqrt(3)/2, 1/2])

rho = DensityMatrix(psi).data

phi = np.pi / 2
Uz = np.array([[np.exp(-1j*phi/2), 0],
              [0, np.exp(1j*phi/2)]])

theta = np.pi / 2
Uy = np.array([[np.cos(theta/2), -1j * np.sin(theta/2)],
              [-1j * np.sin(theta/2), np.cos(theta/2)]])

U = Uz @ Uy
print(U)

rho_rot = U @ rho @ U.conj().T
# Bloch vector after rotation
x = np.trace(rho_rot @ X).real
y = np.trace(rho_rot @ Y).real
z = np.trace(rho_rot @ Z).real

bloch_vector = [x, y, z]
print(bloch_vector)

# Create Bloch sphere manually
b = Bloch()
b.add_vectors(bloch_vector)
b.title = "Bloch vector after rotation about y-axis and z-axis"
b.show()
```
